```python
import glob
from tqdm import tqdm
```

```python
files = glob.glob(r'portale\*.json')

len(files)
```

```python
df =pd.DataFrame()

for f in tqdm(files):
    jf = open(f)

    data=json.loads(jf.read())
    gdf = gpd.GeoDataFrame(data)
    df_this = pd.json_normalize(gdf["features"])
    
    df_this['portal']= f.split('_')[-1].replace('.json','')
    
    df=df.append(df_this)
```

```python
df
```

```python
from shapely.geometry import Point
df['Coordinates'] = df['geometry.coordinates'].apply(Point)
```

```python
gdf_ = gpd.GeoDataFrame(df, geometry='Coordinates', crs='EPSG:4326')
```

```python
gdf_.plot()
```

```python
gdf_
```

```python
gdf_.plot(kind='scatter', y='properties.distance_overtaker', x='properties.distance_stationary', alpha=0.1)
```

```python
gdf_['properties.distance_overtaker']=gdf_['properties.distance_overtaker']*100
gdf_['properties.distance_stationary']=gdf_['properties.distance_stationary']*100
```

```python
import matplotlib.pyplot as plt
```

```python
data_c_sample=gdf_[(gdf_['properties.distance_stationary']>10)& (gdf_['properties.distance_overtaker']>10) ].copy()
```

```python
data_c_sample.plot(kind='scatter', y='properties.distance_overtaker', x='properties.distance_stationary', alpha=0.1)
```

```python
bins = [0, 20, 40, 60, 80, 100, 120, 140, 160, 180, 200]
data_c_sample['Right_binned'] = pd.cut(data_c_sample['properties.distance_stationary'], bins)
```

```python
data_c_sample.boxplot(column=['properties.distance_overtaker'], by=['Right_binned'], figsize=(15,10))
```

```python
ax=data_c_sample.boxplot(column=['properties.distance_overtaker'], by=['Right_binned'] ,figsize=(15,7)) \
        .set(
            axisbelow=True,  # Hide the grid behind plot objects
            #title='Gemessene Geschwindigkeiten je geltende Geschwindigkeitsbegrenzeung \n (Quelle: Envirocar FC-Daten, 2013-2021)',
            #adj = 0,
            #line=-1,
            #suptitle='',
            ylabel='Abstand links (Überholer)',
            xlabel='Abstand rechts (Fahrbahnrand)',
          #  xticklabels=['50','60','70','80','90','100','120','130','kein Tempolimit']
        )
#plt.suptitle('')
#plt.title('Überholabstand (links) in Abhängigkeit von Abstand (rechts) zum Fahrbahnrand', pad=40)
plt.title('Überholabstand (links) in Abhängigkeit von Abstand (rechts) zum Fahrbahnrand', pad=40, y=1.05, fontsize=16)
plt.suptitle('Quelle: OBS-Portale Berlin-Brandenburg, Essen, Darmstadt, BW und OBS-Testportal ', y=1.00, fontsize=12)
#plt.set_title('te')

medians=data_c_sample.groupby('Right_binned')['properties.distance_overtaker'].median().reset_index()

for index, row in medians.iterrows():
    #print (index)
    #print (row['properties.phenomenons.Speed.value'])
    plt.text(index+1, row['properties.distance_overtaker']+2, 
             ("median:\n" + str(round(row['properties.distance_overtaker'],1))), 
             #bbox=dict(facecolor='green', 
             #          alpha=0.5), 
             ha='center',
             color='green',
             fontsize=10)
    
cnts=data_c_sample.groupby('Right_binned')['properties.distance_overtaker'].count().reset_index()

for index, row in cnts.iterrows():
    #print (index)
    #print (row['properties.phenomenons.Speed.value'])
    plt.text(index+1, 320, 
             ("n=" + str(row['properties.distance_overtaker'])), 
             bbox=dict(facecolor='green', 
                       alpha=0.1), 
             #dy=0,
             ha='center',
             color='gray',
             fontsize=10)


plt.show()
```
